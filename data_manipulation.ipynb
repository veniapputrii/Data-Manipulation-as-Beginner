# MANIPULATION
1. Join date & Start time
2. Join date & end time
3. String to date & time
4. Count minutes

```python
# Create a DataFrame from the previous data
mnp = pd.DataFrame(join2)
```

a. convert "Start" column to string data type
b. Joining Date and Start columns

```python
dst = pd.DataFrame(mnp)
# Convert 'Start' column to string
dst['Start'] = dst['Start'].astype(str)
# Join Date and Start columns
dst['DateStartTime'] = dst['Date'] + 'T' + dst['Start']
dst
```

c. Convert column "End" to string data type
d. Joining date and End columns

```python
det = pd.DataFrame(mnp)
#convert 'End' column to string
det['End'] = det['End'].astype(str)
# Join Date and End columns
det['DateEndTime'] = det['Date'] + 'T' + det['End']
det
```

e. convert data type from string to datetime 

```python
from datetime import datetime

# Create DataFrame
mnp = pd.DataFrame(det)

# Convert strings to datetime objects -> DateStartTime's column & DateEndTime's Column
mnp[['DateStartTime', 'DateEndTime']] = mnp[['DateStartTime', 'DateEndTime']].apply(pd.to_datetime)

#Show data types
data_types = mnp.dtypes
data_types
```

f. Time granularity

```python
# Calculate the difference in minutes
mnp['granularity'] = (mnp['DateEndTime'] - mnp['DateStartTime']).dt.total_seconds() / 60
mnp
```
